## Naive Bayes

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.naive_bayes import GaussianNB

centers_arr = [[2., 2.],
               [5., 5.],
               [8., 8.]]

x, labels = make_blobs(n_samples = 30, centers=centers_arr, n_features=2, 
                  random_state=3,
                  shuffle=False,          # false,      (     )
                  cluster_std=1.5)        #standard deviation   (?)  

#print (x)
#print (labels)

plt.rcParams["figure.figsize"] = [12, 8]
plt.scatter(x[:, 0], x[:, 1], c=labels, s=50)


naiveBayes = GaussianNB()
naiveBayes.fit(x, labels)


rng = np.random.RandomState(0)
xnew = [-2.0, -2.0] + [12, 12] * rng.rand(2000, 2)
ynew = naiveBayes.predict(xnew)

plt.scatter(xnew[:, 0], xnew[:, 1], c=ynew, s=10, alpha=0.2)


xnew2 = [-2.0, -2.0] + [12, 12] * rng.rand(20, 2)
ynew2 = naiveBayes.predict(xnew)
y_proba = naiveBayes.predict_proba (xnew2)
print (y_proba)
```
